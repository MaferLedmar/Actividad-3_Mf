```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Ventas_totales.csv')
data
```

**Limpieza de datos (nulos)**

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data2=data.copy()
```

```python
#TARJETAS DE CREDITO
#Sustituir valores nulos con promedio o media
data2["tarjetas_credito"]=data2["tarjetas_credito"].fillna(round(data["tarjetas_credito"].mean(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#TARJETAS DE DEBITO
#Sustituir valores nulos con mediana
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(data["tarjetas_debito"].median(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#subtotal_ventas_alimentos_bebidas - 10
#salon_ventas - 6

#Sustituir valores nulos con promedio o media
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))

data2["salon_ventas"]=data2["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#otros_medios -1
#otros - 4

#Sustituir valores nulos con mediana
data2["otros_medios"]=data2["otros_medios"].fillna(round(data["otros_medios"].median(),1))
data2["otros"]=data2["otros"].fillna(round(data["otros"].median(),1))

#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#bebidas - almacen - panaderia - lacteos - carnes 
#Sustituir valores nulos por valores no nulos hacia adelante forward fill" ("ffill")

data2["bebidas"] =data2["bebidas"].fillna(method="ffill") 
data2["almacen"] =data2["almacen"].fillna(method="ffill") 
data2["panaderia"] =data2["panaderia"].fillna(method="ffill") 
data2["lacteos"] =data2["lacteos"].fillna(method="ffill") 
data2["carnes"] =data2["carnes"].fillna(method="ffill") 

data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#verduleria_fruteria - alimentos_preparados_rotiseria - indumentaria_calzado_textiles_hogar
# electronicos_articulos_hogar

#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")

data2["verduleria_fruteria"] =data2["verduleria_fruteria"].fillna(method="bfill") 
data2["alimentos_preparados_rotiseria"] =data2["alimentos_preparados_rotiseria"].fillna(method="bfill") 
data2["indumentaria_calzado_textiles_hogar"] =data2["indumentaria_calzado_textiles_hogar"].fillna(method="bfill") 
data2["electronicos_articulos_hogar"] =data2["electronicos_articulos_hogar"].fillna(method="bfill") 

data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Obtenemos la cuenta de los elementos nulos del dataframe completo
Cuenta_Total_Nulos= data2.isnull().sum().sum()
Cuenta_Total_Nulos
```

***Outliers***

Revisamos las variables y ocupamos el **método IQR**

```python
#Realizamos diagrama de caja o bigote
#VENTAS_PRECIOS_CORRIENTES
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data2["ventas_precios_corrientes"]) 
plt.title("ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr["ventas_precios_corrientes"]) 
plt.title("ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```
